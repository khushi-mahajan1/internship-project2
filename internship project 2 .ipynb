```python
# task No 2
import torch.nn as nn
import torch.nn.functional as F
from torchvision.utils import save_image
from torch.nn.utils import spectral_norm

class TextEncoder(nn.Module):
  def __init__(self,vocab_size,embed_dim):
    super(TextEncoder,self).__init__()
    self.embedding=nn.Embedding(vocab_size,embed_dim)

  def forward(self,captions):
    return self.embedding(captions).mean(dim=1)

class SelfAttention(nn.Module):
  def __init__(self,in_dim):
    super(SelfAttention,self).__init__()
    self.query=nn.Conv2d(in_dim,in_dim//8,1)
    self.key=nn.Conv2d(in_dim,in_dim//8,1)
    self.value=nn.Conv2d(in_dim,in_dim,1)
    self.gamma=nn.Parameter(torch.zeros(1))

  def forward(self,x):
    batch,C,W,H=x.size()
    proj_query=self.query(x).view(batch,-1,W*H).permute(0,2,1)
    proj_key=self.key(x).view(batch,-1,W*H)
    attention=F.softmax(torch.bmm(proj_query,proj_key),dim=-1)
    proj_value=self.value(x).view(batch,-1,W*H)
    out=torch.bmm(proj_value,attention.permute(0,2,1))
    out=out.view(batch,C,W,H)
    return self.gamma * out + x

class Generator(nn.Module):
  def __init__(self,z_dim,text_dim,img_channels=3):
    super(Generator,self).__init__()
    self.fc=nn.Linear(z_dim + text_dim,256 * 4 * 4)

    self.Conv_blocks=nn.Sequential(
        nn.BatchNorm2d(256),
        nn.Upsample(scale_factor=2),
        nn.Conv2d(256,256,3,padding=1),
        nn.BatchNorm2d(256),
        nn.ReLU(),
        SelfAttention(256),
        nn.Upsample(scale_factor=2),
        nn.Conv2d(256,128,3,padding=1),
        nn.BatchNorm2d(128),
        nn.ReLU(),

        nn.Upsample(scale_factor=2),
        nn.Conv2d(128,64,3,padding=1),
        nn.BatchNorm2d(64),
        nn.ReLU(),

        nn.Upsample(scale_factor=2),
        nn.Conv2d(64,img_channels,3,padding=1),
        nn.Tanh()
    )

  def forward(self,z,text_embed):
    x=torch.cat([z,text_embed],dim=1)
    x=self.fc(x).view(-1,256,4,4)
    img=self.Conv_blocks(x)
    return img

class Discriminator(nn.Module):
  def __init__(self,img_channels=3,text_dim=100):
    super(Discriminator,self).__init__()
    self.Conv=nn.Sequential(
        spectral_norm(nn.Conv2d(img_channels,64,4,stride=2,padding=1)),
        nn.LeakyReLU(0.2,inplace=True),

        spectral_norm(nn.Conv2d(64,128,4,stride=2,padding=1)),
        nn.LeakyReLU(0.2,inplace=True),

        spectral_norm(nn.Conv2d(128,256,4,stride=2,padding=1)),
        nn.LeakyReLU(0.2,inplace=True),

        SelfAttention(256),
    )
    self.fc=nn.Linear(256*8*8 + text_dim,1)

  def forward(self,img,text_embed):
    x=self.Conv(img)
    x=x.view(x.size(0),-1)
    x=torch.cat([x,text_embed],dim=1)
    validity=torch.sigmoid(self.fc(x))
    return validity

batch_size=4
z_dim=100
text_dim=50
vocab_size=5000
seq_len=10

z=torch.randn(batch_size,z_dim)
captions=torch.randint(0,vocab_size,(batch_size,seq_len))
text_encoder=TextEncoder(vocab_size,text_dim)
text_embed=text_encoder(captions)
gen=Generator(z_dim,text_dim)
disc=Discriminator(text_dim=text_dim)
fake_images=gen(z,text_embed)
validity=disc(fake_images,text_embed)
save_image(fake_images,'sample64.png',nrow=2,normalize=True)
print("Generated 64*64 images saved as sample64.png")
from IPython.display import Image,display
display(Image('sample64.png'))
```

```python
# task no 5
# Install required libraries
# pip install torch transformers

import torch  # task no 5
from transformers import CLIPTokenizer, CLIPTextModel

# -----------------------------
# 1. Load CLIP tokenizer and text model
# -----------------------------
tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-base-patch32")
text_model = CLIPTextModel.from_pretrained("openai/clip-vit-base-patch32")

# -----------------------------
# 2. Define a function to convert text to embeddings
# -----------------------------
def get_text_embedding(text_list, device="cpu"):
    """
    Converts a list of text descriptions to CLIP embeddings
    Args:
        text_list: List of strings (descriptions)
        device: 'cpu' or 'cuda'
    Returns:
        Tensor of shape [batch_size, embedding_dim]
    """
    # Tokenize the text
    tokens = tokenizer(text_list, padding=True, truncation=True, return_tensors="pt")
    tokens = {k: v.to(device) for k, v in tokens.items()}

    # Pass tokens through the text model to get embeddings
    with torch.no_grad():
        outputs = text_model(**tokens)
        # Take mean of the last hidden state across sequence length to get fixed-size embeddings
        embeddings = outputs.last_hidden_state.mean(dim=1)

    return embeddings

# -----------------------------
# 3. Example usage
# -----------------------------
text_descriptions = [
    "A beautiful sunset over the mountains",
    "A cute dog playing in the park",
    "A futuristic cityscape at night"
]

device = "cuda" if torch.cuda.is_available() else "cpu"
text_model.to(device)

# Get embeddings
embeddings = get_text_embedding(text_descriptions, device=device)
print("Embeddings shape:", embeddings.shape)

```

```python
# task no 6

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt
from skimage.draw import disk, polygon

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Parameters
image_size = 32
latent_dim = 100
num_classes = 3  # circle, square, triangle
channels = 1

# Dataset for shapes
class ShapeDataset(Dataset):
    def __init__(self, n=3000):
        self.images, self.labels = [], []
        for _ in range(n):
            label = np.random.randint(0, num_classes)
            img = np.zeros((image_size, image_size), dtype=np.float32)
            if label == 0:  # Circle
                rr, cc = disk((image_size//2, image_size//2), image_size//3)
                img[rr, cc] = 1.0
            elif label == 1:  # Square
                start = image_size//4
                end = image_size - start
                img[start:end, start:end] = 1.0
            else:  # Triangle
                r = np.array([8, 24, 24])
                c = np.array([16, 8, 24])
                rr, cc = polygon(r, c)
                img[rr, cc] = 1.0
            img = img * 2 - 1  # scale to [-1, 1]
            self.images.append(img[None, :, :])
            self.labels.append(label)

    def __len__(self): return len(self.images)
    def __getitem__(self, idx):
        return torch.tensor(self.images[idx], dtype=torch.float32), torch.tensor(self.labels[idx])

dataset = ShapeDataset()
dataloader = DataLoader(dataset, batch_size=64, shuffle=True)

# Generator
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, num_classes)
        self.init_size = image_size // 4
        self.fc = nn.Sequential(
            nn.Linear(latent_dim + num_classes, 256 * self.init_size**2),
            nn.BatchNorm1d(256 * self.init_size**2),
            nn.ReLU(True)
        )
        self.conv = nn.Sequential(
            nn.ConvTranspose2d(256, 128, 4, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 4, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.Conv2d(64, channels, 3, stride=1, padding=1),
            nn.Tanh()
        )

    def forward(self, z, labels):
        c = self.label_emb(labels)
        x = torch.cat([z, c], 1)
        out = self.fc(x)
        out = out.view(out.size(0), 256, self.init_size, self.init_size)
        img = self.conv(out)
        return img

# Discriminator
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, num_classes)
        self.conv = nn.Sequential(
            nn.Conv2d(channels + 1, 64, 4, stride=2, padding=1),
            nn.LeakyReLU(0.2),
            nn.Conv2d(64, 128, 4, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2),
            nn.Conv2d(128, 256, 3, stride=1, padding=1),
            nn.LeakyReLU(0.2),
            nn.Flatten(),
            nn.Linear(256 * (image_size//4) * (image_size//4), 1),
            nn.Sigmoid()
        )

    def forward(self, img, labels):
        c = self.label_emb(labels).unsqueeze(2).unsqueeze(3)
        c = c.repeat(1, 1, image_size, image_size)
        x = torch.cat([img, c[:, :1, :, :]], 1)
        return self.conv(x)

# Models
generator = Generator().to(device)
discriminator = Discriminator().to(device)

# Loss + Optimizers
adversarial_loss = nn.BCELoss()
opt_G = optim.Adam(generator.parameters(), lr=0.0002, betas=(0.5, 0.999))
opt_D = optim.Adam(discriminator.parameters(), lr=0.0002, betas=(0.5, 0.999))

# Training
epochs = 20
for epoch in range(epochs):
    for imgs, labels in dataloader:
        imgs, labels = imgs.to(device), labels.to(device)
        bs = imgs.size(0)
        valid = torch.ones(bs, 1, device=device)
        fake = torch.zeros(bs, 1, device=device)

        # Train Generator
        opt_G.zero_grad()
        z = torch.randn(bs, latent_dim, device=device)
        gen_labels = torch.randint(0, num_classes, (bs,), device=device)
        gen_imgs = generator(z, gen_labels)
        g_loss = adversarial_loss(discriminator(gen_imgs, gen_labels), valid)
        g_loss.backward()
        opt_G.step()

        # Train Discriminator
        opt_D.zero_grad()
        real_loss = adversarial_loss(discriminator(imgs, labels), valid)
        fake_loss = adversarial_loss(discriminator(gen_imgs.detach(), gen_labels), fake)
        d_loss = (real_loss + fake_loss) / 2
        d_loss.backward()
        opt_D.step()

    print(f"Epoch [{epoch+1}/{epochs}] | D Loss: {d_loss.item():.4f} | G Loss: {g_loss.item():.4f}")

# Test generation
generator.eval()
z = torch.randn(9, latent_dim, device=device)
sample_labels = torch.tensor([0,1,2,0,1,2,0,1,2], device=device)
gen_imgs = generator(z, sample_labels).cpu().detach()

fig, axs = plt.subplots(3, 3, figsize=(6,6))
for i, ax in enumerate(axs.flatten()):
    img = (gen_imgs[i][0] + 1) / 2  # scale back to [0,1]
    ax.imshow(img, cmap="gray")
    ax.set_title(["Circle","Square","Triangle"][sample_labels[i]])
    ax.axis("off")
plt.tight_layout()
plt.show()
```
